```python
qs['date'] = pd.to_datetime(qs['AskedOnUtc'])
qs['month_year'] = qs['date'].dt.month.astype(str) + '-' + qs['date'].dt.year.astype(str)
#qs['month_year'] = pd.to_datetime(qs['month_year'])
ts_counts = qs.groupby(['month_year', 'Category']).size().unstack(fill_value=0).reset_index()
#chart
#counts.columns.name = None
#counts.rename(columns={'month_year': 'date'}, inplace=True)

# ts_counts.sort_values(by="month_year") # un sort it
```

```python
alt.Chart(ts_counts).mark_line().encode(
    x=alt.X('month_year'),
    y='Family and Children'
)

# Where are clients asking questions?
# See demographics of people asking questions vs demographics of that zip code
```

```python
# Create a list of charts to concatenate
facet_cols = list(ts_counts.columns)[1:]
charts = []

for cat in facet_cols:
    chart = alt.Chart(ts_counts).mark_line().encode(
        x='month_year',
        y=cat
    ).properties(
        width=200,
        height=200
    )
    charts.append(chart)
    
# Concatenate the charts into a facet chart
facet_chart = charts[0]
facet_chart
for chart in charts[1:]:
    facet_chart &= chart

# Configure the chart
facet_chart = facet_chart.properties(
    title='Faceted Charts'
).configure_axis(
    labelFontSize=5,
    titleFontSize=14
).configure_title(
    fontSize=16,
    fontWeight='bold',
    anchor='middle'
)

facet_chart
```
